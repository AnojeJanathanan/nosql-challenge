5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
from decimal import Decimal

d = establishments.find({"longitude": {"$type": "string"}, "latitude": {"$type": "string"}})

for document in d:
    longitude = Decimal(document['longitude'])
    latitude = Decimal(document['latitude'])
    
    establishments.update_many(
        {'_id': document['_id']},
        {'$set': {'longitude': longitude, 'latitude': latitude}}
    )
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$not": {"$in": [1, 2, 3, 4, 5]}}}, {"$set": {"RatingValue": None}})
```

```python
# Change the data type from String to Integer for RatingValue

rating_val = establishments.find({"RatingValue": {"$type": "string"}})

for document in rating_val:
    rating_value = int(document['RatingValue'])
    
    establishments.update_one(
        {'_id': document['_id']},
        {'$set': {'RatingValue': rating_value}}
    )

```
